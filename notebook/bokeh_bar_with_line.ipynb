```python
from bokeh.charts import Bar, output_notebook, show
from bokeh.models.ranges import Range1d

from bokeh.models import ColumnDataSource
from bokeh.models.glyphs import Line
import numpy as np

# create dummy data
df = dict(
    x=[1, 2, 3, 4, 5],
    y=[6, 7, 6, 4, 5],
    h=[2, 8, 5, 10, 7]
)

p = Bar(data=df, label='x', values='h', title="Bar and Line Plot")

source = ColumnDataSource(data=df)
glyph = Line(x='x', y='y', line_color='grey', line_width=2)
p.add_glyph(source, glyph)

# Setting the y  axis range   
p.y_range = Range1d(0, 12)

output_notebook()
show(p)
```

```python
import pandas, os
from bokeh.charts import Bar, output_notebook, show, Line
from bokeh.layouts import column, row

csv_filepath = ('big_data_benchmarking_20170125.csv')
dataframe = pandas.read_csv(csv_filepath, low_memory=False)
dataframe = dataframe[dataframe['concurrency_factor'] == 1]
source = ColumnDataSource(data=dataframe)
# Bar Plot
bar_query_id_by_database = Bar(source.to_df(), label='query_id', values='time', group='database',
                               legend='top_right', xlabel='Query ID', ylabel='Time in seconds',
                               title="Concurrency Avg execution time by query id", agg='mean')
# Line Plot
avg_query_time_dataframe = source.to_df().pivot_table(index=['query_id', 'name'], columns='database',
                                                      values='time', aggfunc='mean').fillna('')
database_columns = avg_query_time_dataframe.columns
print([c for c in database_columns])
avg_query_time_dataframe.reset_index(inplace=True)
new_cols = avg_query_time_dataframe.columns
print([c for c in new_cols])
line_query_id_by_database = Line(avg_query_time_dataframe, x='query_id', y=[col for col in database_columns],
                                 xlabel='Query ID', ylabel='Time in seconds',
                                 color=[col for col in database_columns], legend='top_right',
                                 legend_sort_field='color',
                                 title="Concurrency Avg execution time by query id", 
                                 tooltips=[('Query ID', '@query_id'), ('Time', '$y{0.000}'), ('Database', "@{Oracle Database}")])
show(line_query_id_by_database)
```
